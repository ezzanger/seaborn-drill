```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import xlrd
tips = sns.load_dataset("tips")

print(tips.columns)
```

# Tips by Smokers (and by gender)


```python


# Creating variables for each of the four conditions to graph using the basic
# matplotlib functionality.
mensmoker = tips.loc[(tips['sex']=='Male')&(tips['smoker']=='Yes'),'tip']
womensmoker = tips.loc[(tips['sex']=='Female')&(tips['smoker']=='Yes'),'tip']
mennonsmoker = tips.loc[(tips['sex']=='Male')&(tips['smoker']=='No'),'tip']
womennonsmoker = tips.loc[(tips['sex']=='Female')&(tips['smoker']=='No'),'tip']

# Making overlapping histograms.
plt.hist(mensmoker, color='red',  alpha=.5,label='Men Smoker Tip')  # alpha just controls the opacity
plt.hist(womensmoker, color='blue', alpha=.5,label='Women Smoker Tip')
plt.hist(mennonsmoker, color='green', alpha=.5,label='Men Non Smoker Tip')  
plt.hist(womennonsmoker, color='yellow', alpha=.5,label='Women Non Smoker Tip')
plt.xlabel('Tip in USD')
plt.legend(loc='upper right')
plt.title('Plot 1: Meal Tips by Gender and Smoker')
plt.show()

```

```python

sns.set(style="ticks") 
#heading
g = sns.FacetGrid(tips, row="sex", col="smoker")
g.map(plt.hist, "tip", color="red",  lw=0)
plt.subplots_adjust(top=0.9)
# Making a more informative axis name.
g.set_axis_labels('Tip in USD')
plt.suptitle('Plot 2: Meal Tips by Gender and Smoking')
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()
```

```python

ax = sns.boxplot(x='smoker',y='tip',hue='sex',data=tips)  
plt.title('Plot 3: Meal Tips by Gender and Smoker')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tips')

plt.show()
```

```python
smokey = (tips['smoker']== 'Yes')
smoken = (tips['smoker']== 'No')

plt.plot(smokey, tips['tip'], label= "Smoker Tips")
plt.plot(smoken, tips['tip'], label= "Non Smoker Tips")
plt.show()
```

# Tips based on Total Bill


```python

sns.set(style="darkgrid")

g = sns.factorplot(x="total_bill", y="tip", hue="time", data=tips,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Tips")
g.set_xlabels("")
plt.title('Barplot: Meal Tips by Total Bill')
plt.show()


```

```python
g = sns.lmplot(y='tip', # Variable 1.
               x='total_bill', # Variable 2.
               data=tips, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Tips")
g.set_xlabels("Total Bill")
plt.title('Scatterplot: Meal Tips by Total Bill')
plt.show()
```

```python
ax = sns.pointplot(x="total_bill", y="tip", data=tips, capsize=.2)
```

```python

sns.set(style="ticks") 
#heading
g = sns.FacetGrid(tips, row="time", col="tip")
g.map(plt.hist, "tip", color="red",  lw=0)
plt.subplots_adjust(top=0.9)
# Making a more informative axis name.
g.set_axis_labels('Tip in USD')
plt.suptitle('Plot 2: Meal Tips by Meal')
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()
```

# Tips based on Meal (lunch/dinner)
included is whether the guest was a smoker or not

```python

ax = sns.boxplot(x='time',y='tip',hue='smoker',data=tips)  
plt.title('Plot 3: Meal Tips by Smoker and Time')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tips')

plt.show()
```

```python
sns.set(style="darkgrid")

g = sns.factorplot(x="time", y="tip", hue="smoker", data=tips,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Tips")
g.set_xlabels("")
plt.title('Barplot: Meal Tips by Time and Gender')
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set(style="white")

g = sns.lmplot(y='tip', 
               x='total_bill',
               hue='smoker',
               data=tips, # Data.
               col='time',
               fit_reg=True,
               ci=False,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Tips")
g.set_xlabels("Total Bill")
plt.suptitle('Scatterplot Regression: Meal Tips by Total Bill, split by gender and time')
plt.subplots_adjust(top=0.9)
plt.show()
```

```python

# Setting the overall aesthetic.
sns.set(style="whitegrid")

ax = sns.boxplot(x='time',y='tip',hue='smoker',data=tips,palette='pastel')  

plt.title('Plot 4: Meal Tips by Time and Gender')

sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tips')

plt.show()
```

```python

sns.set(style="ticks") 
#heading
g = sns.FacetGrid(tips, row="time", col="smoker")
g.map(plt.hist, "tip", color="red",  lw=0)
plt.subplots_adjust(top=0.9)
# Making a more informative axis name.
g.set_axis_labels('Tip in USD')
plt.suptitle('Plot 2: Meal Tips by Gender and Smoking')
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set(style="whitegrid")

g = sns.factorplot(x="time", y="tip", hue="sex", data=tips,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Tips")
g.set_xlabels("")
plt.title('Pointplot: Meal Tips by Time and Smoker')
plt.show()
```
